```python
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

# Load Images
images = {
    "train":{},
    "test":{}
}
categories = ["people", "car", "street sign", "bus", "motorcycle"]
for typ in ["train","test"]:
    for category in categories:
        files = os.listdir(f"../data_set/{typ}/{category}")
        for file in files:
            path = f"../data_set/{typ}/{category}/{file}"
            if path.endswith(".jpg") or path.endswith(".jpeg"):
                img = cv2.imread(path)
                aux = img
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                images[typ][category] = images[typ].get(category,[]) + [img]

img = cv2.imread("data_set\test\bus\000000003125.jpg")
```

```python
objects_gray = {
    "train":{},
    "test":{}
}
categories = ["people", "car", "street sign", "bus", "motorcycle"]
for typ in ["train", "test"]:
    for category in categories:
        files = os.listdir(f"../data_set/{typ}/{category}")
        for file in files:
            path = f"../data_set/{typ}/{category}/{file}"
            if path.endswith(".jpg") or path.endswith(".jpeg"):
                img = cv2.imread(path)
                gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
                objects_gray[typ][category] = objects_gray[typ].get(category, []) + [gray_img]
```
